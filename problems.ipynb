## Problem 1

In problem, you will solve a baby version of a famous type of mathematical equations - Pell's equations. Pell's equations has the following form:
$$x^2 - dy^2 = 1$$
where $x, y \in \mathbb{Z}$ (they are *integers*!) and $d$ a positive *square-free* integer (i.e., it's not a perfect square). The existence of solutions is one of the foundamental results in elementary number theory. It can be mathematically proved that, either with continued fractions or the more powerful quadratic rings $\mathbb{Z}[d]$, for all positive *square-free* integer $d$, the equation 
$$x^2 - dy^2 = 1$$
has **infinitely** many solutions. Furthermore, the solutions are all generated by a single solution, denoted by $\varepsilon_d$ (meaning for all pairs of solutions when regarded as complex numbers can be written as $\varepsilon_d^n$ for some n). Of course, you will not be asked to prove the existence. But you will need verify some results.

### a) 
Verify that if $(19, 3)$ is a solution to the equation
$$x^2-40y^2=1$$
then $(721, 114)$ is another solution to the equation.

```python
# Here is the given solution. Verify the other one. Press `shift + enter` to run the code
x = 19
y = 3
print(x ** 2 - 40 * y ** 2) # output: 1
```
